## Scrapping de la clasificación general de cada temporada desde 2011 hasta 2023

Se han utilizado las librerías BeautifulSoup y Selenium para automatizar la descarga de datos de clasificación de la NBA entre los años 2011 y 2023.

Después se guardan en un DataFrame y se exportan a .csv

```python
from bs4 import BeautifulSoup as bs
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
```

```python
import numpy as np
import pandas as pd
import zipfile
import requests
import json
```

```python
season_list = []

for n in range(2023, 2010, -1):
    for i in range(30):
        season_list.append(n)

print(len(season_list))
season_list
```

```python
conference_list = []

for n in range(2023, 2010, -1):
    for i in range(15):

        conference_list.append("East")
    
    for j in range(15):
        conference_list.append("West")

print(len(conference_list))
conference_list
```

```python
driver = webdriver.Chrome()

season_num = ["2022-23", "2021-22", "2020-21", "2019-20", "2018-19", "2017-18", "2016-17", "2015-16", "2014-15", "2013-14", "2012-13", "2011-12", "2010-11"]

ranking_list = []
team_city_list = []
lista_wins_tot = []
team_loses_tot_list = []
team_w_perc_list = []
team_rec_loc_list = []
team_rec_vis_list = []



for season in season_num:
    url = "https://www.nba.com/standings?Season=" + season
    print(url)

    driver.get(url)

    # Cookies = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/div[1]/div/div[2]/div/button[1]")
    # Cookies.click()

    soup = bs(driver.page_source,"lxml")

    rankings = soup.select('div.Crom_base__f0niE tr.StatsStandingsTable_row__o6A7G span.StatsStandingsTable_teamRank__dbHMs')

    # ranking_list = []

    for ranking in rankings:
        ranking_list.append(ranking.get_text())

    ranking_list

    # team_city_list = []

    team_cities = soup.select('div.Crom_base__f0niE tr.StatsStandingsTable_row__o6A7G span.StatsStandingsTable_teamCity__WJIh9')

    for team_city in team_cities:
        team_city_list.append(team_city.get_text().replace("\xa0", ""))

    team_city_list

    team_wins = soup.select('div.Crom_base__f0niE tr.StatsStandingsTable_row__o6A7G td')

    team_wins_tot_list = []

    for team_win in team_wins:
        team_wins_tot_list.append(team_win.get_text())

    lista_info_equipos = []

    for i in range(0, 360, 12):
        info_equipo = team_wins_tot_list[i:i + 12]
        lista_info_equipos.append(info_equipo)

    lista_info_equipos

    # lista_wins_tot = []
    # team_loses_tot_list = []
    # team_w_perc_list = []
    # team_rec_loc_list = []
    # team_rec_vis_list = []

    for equipo in lista_info_equipos:
        lista_wins_tot.append(equipo[1])
        team_loses_tot_list.append(equipo[2])
        team_w_perc_list.append(equipo[3])
        team_rec_loc_list.append(equipo[7])
        team_rec_vis_list.append(equipo[8])

    # print(len(team_city_list))
    # print(len(lista_wins_tot))
    # print(len(team_loses_tot_list))
    # print(len(team_w_perc_list))
    # print(len(team_rec_loc_list))
    # print(len(team_rec_vis_list))



    team_loses_tot_list

    # driver.close()


print(len(team_city_list))
print(len(lista_wins_tot))
print(len(team_loses_tot_list))
print(len(team_w_perc_list))
print(len(team_rec_loc_list))
print(len(team_rec_vis_list))














    
```

```python
df_standings = pd.DataFrame({"season":season_list,
                                  "conference":conference_list,
                                  "team_name":team_city_list,
                                  "team_wins_tot":lista_wins_tot, 
                                  "team_loses":team_loses_tot_list,
                                  "team_win_perc":team_w_perc_list,
                                  "team_rec_loc":team_rec_loc_list, 
                                  "team_rec_vis":team_rec_vis_list})

df_standings
print(df_standings.to_string())
```
