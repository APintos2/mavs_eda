```python
url = "https://www.nba.com/stats/teams/traditional?sort=W_PCT&dir=-1&SeasonType=Regular+Season&Season=2010-11"

driver.get(url)

soup = bs(driver.page_source,"lxml")

team_info_list = []

team_infos = soup.select('table.Crom_table__p1iZz tbody.Crom_body__UYOcU tr td')

# print(team_infos)

for team_info in team_infos:
    team_info_list.append(team_info.get_text())

print(len(team_info_list))
team_info_list

lista_info_equipos = []

for i in range(0, 840, 28):
    info_equipo = team_info_list[i:i + 28]
    lista_info_equipos.append(info_equipo)


print(lista_info_equipos)

for equipo in lista_info_equipos:
    ranking_list.append(equipo[0])
    team_name_list.append(equipo[1])
    games_played_list.append(equipo[2])
    wins_list.append(equipo[3])
    losses_list.append(equipo[4])
    win_perc_list.append(equipo[5])
    ppg_list.append(equipo[7])
    fgm_list.append(equipo[8])
    fga_list.append(equipo[9])
    fg_perc_list.append(equipo[10])
    triples_made_list.append(equipo[11])
    triples_att_list.append(equipo[12])
    triples_perc_list.append(equipo[13])
    ft_made_list.append(equipo[14])
    ft_att_list.append(equipo[15])
    ft_perc_list.append(equipo[16])
    off_reb_list.append(equipo[17])
    def_reb_list.append(equipo[18])
    ast_list.append(equipo[20])
    turnovers_list.append(equipo[21])
    steals_list.append(equipo[22])
    blocks_list.append(equipo[23])
    pf_list.append(equipo[25])
    plus_minus_list.append(equipo[27])

print(len(plus_minus_list))
```

```python
df_team_stats = pd.DataFrame({"season":season_list,
                            "ranking":ranking_list,
                            "team_name":team_name_list,
                            "games_played":games_played_list,
                            "wins":wins_list,
                            "losses":losses_list,
                            "team_win_perc":win_perc_list,
                            "ppg":ppg_list,
                            "fgm":fgm_list,
                            "fga":fga_list,
                            "fg_perc":fg_perc_list,
                            "threes_made":triples_made_list,
                            "threes_att":triples_att_list,
                            "threes_perc":triples_perc_list,
                            "ft_made":ft_made_list,
                            "ft_att": ft_att_list,
                            "ft_perc":ft_perc_list,
                            "off_reb":off_reb_list,
                            "def_reb":def_reb_list,
                            "ast":ast_list,
                            "turnovers":turnovers_list,
                            "steals":steals_list,
                            "blocks":blocks_list,
                            "pers_fouls":pf_list,
                            "plus_minus":plus_minus_list})




df_team_stats
print(df_team_stats.to_string())
```

```python
df_team_stats.to_csv(r'../data/team_stats_2011_2023.csv')
```
