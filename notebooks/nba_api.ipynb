```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
# Obtenemos los datos de clasificaci√≥n de 2021

url = "https://api-nba-v1.p.rapidapi.com/standings"



querystring = {"league":"standard","season":"2021"}

headers = {
	"X-RapidAPI-Key": "6bb069ac2cmshb6360f4caf38decp16ed28jsn3bfde7caa6d8",
	"X-RapidAPI-Host": "api-nba-v1.p.rapidapi.com"
}


res = requests.get(url, headers=headers, params=querystring)

res
```

```python
standings_tot = []

standings = json.loads(res.text)
standings_tot.append(standings)

print(standings)
len(standings_tot)
```

```python
season = []
conference = []
team_name = []
team_code = []
team_wins_tot = []
team_wins_loc = []
team_wins_vis = []
team_w_perc = []
team_loses_tot = []


for i in range(len(standings["response"])):
    season.append(standings["response"][i]["season"])
    conference.append(standings["response"][i]["conference"]["name"])
    team_name.append(standings["response"][i]["team"]["name"])
    team_code.append(standings["response"][i]["team"]["code"])
    team_wins_tot.append(standings["response"][i]["win"]["total"])
    team_wins_loc.append(standings["response"][i]["win"]["home"])
    team_wins_vis.append(standings["response"][i]["win"]["away"])
    team_w_perc.append(standings["response"][i]["win"]["percentage"])
    team_loses_tot.append(standings["response"][i]["loss"]["total"])


season
conference
team_loses_tot
```

```python
standings_2011_2021 = pd.DataFrame({"season":season, 
                                    "conference":conference, 
                                    "team_name":team_name, 
                                    "team_code":team_code, 
                                    "team_wins_tot":team_wins_tot, 
                                    "team_wins_loc":team_wins_loc, 
                                    "team_wins_vis":team_wins_vis, 
                                    "team_win_perc":team_w_perc, 
                                    "team_loses":team_loses_tot})

standings_2011_2021
```
