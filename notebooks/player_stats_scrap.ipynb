```python
player_info_list = []

player_infos = soup.select('table.Crom_table__p1iZz tbody.Crom_body__UYOcU tr td')

for player_info in player_infos:
    player_info_list.append(player_info.get_text())

print(len(player_info_list))
player_info_list

lista_info_jugs = []

for i in range(0, 690, 30):
    info_equipo = player_info_list[i:i + 30]
    lista_info_jugs.append(info_equipo)

print(len(lista_info_jugs[0]))
lista_info_jugs
```

```python
player_name_list = []
player_age_list = []
games_played_list = []
wins_list = []
losses_list = []
mins_played_list = []
ppg_list = []
fgm_list = []
fga_list = []
fg_perc_list = []
triples_made_list = []
triples_att_list = []
triples_perc_list = []
ft_made_list = []
ft_att_list = []
ft_perc_list = []
reb_list = []
ast_list = []
turnovers_list = []
steals_list = []
blocks_list = []
pf_list = []
plus_minus_list = []


for equipo in lista_info_jugs:
        player_name_list.append(equipo[1])
        player_age_list.append(equipo[3])
        games_played_list.append(equipo[4])
        wins_list.append(equipo[5])
        losses_list.append(equipo[6])
        mins_played_list.append(equipo[7])
        ppg_list.append(equipo[8])
        fgm_list.append(equipo[9])
        fga_list.append(equipo[10])
        fg_perc_list.append(equipo[11])
        triples_made_list.append(equipo[12])
        triples_att_list.append(equipo[13])
        triples_perc_list.append(equipo[14])
        ft_made_list.append(equipo[15])
        ft_att_list.append(equipo[16])
        ft_perc_list.append(equipo[17])
        reb_list.append(equipo[20])
        ast_list.append(equipo[21])
        turnovers_list.append(equipo[22])
        steals_list.append(equipo[23])
        blocks_list.append(equipo[24])
        pf_list.append(equipo[25])
        plus_minus_list.append(equipo[29])

print(len(player_name_list))
```

```python
df_player_stats = pd.DataFrame({"season":season_list,
                            "player_name":player_name_list,
                            "player_age":player_age_list,
                            "games_played":games_played_list,
                            "wins":wins_list,
                            "losses":losses_list,
                            "mins_played":mins_played_list,
                            "ppg":ppg_list,
                            "fgm":fgm_list,
                            "fga":fga_list,
                            "fg_perc":fg_perc_list,
                            "threes_made":triples_made_list,
                            "threes_att":triples_att_list,
                            "threes_perc":triples_perc_list,
                            "ft_made":ft_made_list,
                            "ft_att": ft_att_list,
                            "ft_perc":ft_perc_list,
                            "reb":reb_list,
                            "ast":ast_list,
                            "turnovers":turnovers_list,
                            "steals":steals_list,
                            "blocks":blocks_list,
                            "pers_fouls":pf_list,
                            "plus_minus":plus_minus_list})

df_player_stats
print(df_player_stats.to_string())
```
